# Beugungs- und Interferenzexperimenten mit Mikrowellen in 2D

Die folgenden Beugungs- und Interferenzexperimente mit Mikrowellen lassen sich in guter Näherung in 2D beschreiben:

- V5: Doppelspalt
- V6: Michelson-Interferometer
- V7: Wellenausbreitung in Acrylglas

Für diese Experimente finden Sie Anleitungen zu einem möglichen Aufbau des Experiments, der Durchführung und der Auswertung unter [`XLAB/2-Schwingungen/2.4-Welle2D/Experiment/`](\.).
