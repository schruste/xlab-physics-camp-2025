```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
t = np.pi / 2
a = 2 * np.pi
c_1 = 1
c_2 = 1
b = 4 * np.pi
w = 1
```

```python
@np.vectorize
def u(x):
    k_1 = w/c_1
    k_2 = w/c_2
    if(x <= a):
        return np.cos(w * t - k_1 * x)
    else:
        return np.cos(w * t - k_2 * x + (k_2 - k_1)*a)
```

```python
x = np.linspace(0, b, 1000)
plt.plot(x, u(x))
c_2 = 2/3
plt.plot(x, u(x))
```
